### Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sns
```

```python
f = open('logins.json', 'r')
data = json.loads(f.read())
```

```python
# convert login json data to dataframe

df_logins = pd.DataFrame(data)
df_logins.head()
```

```python
# Find the max and min of the datetime index

max(df_logins.index), min(df_logins.index)
```

**Convert login_time to datetime data type**

```python
df_logins['login_time'] = pd.to_datetime(df_logins['login_time'], infer_datetime_format=True)
```

```python
df_logins.info()
```

```python
df_logins.head()
```

**Since we need to create 15 minute intervals goupings, we might as well create groupings for day, month and year as it could potentially lead to additional insights**

```python
# Create time features

df_logins['year'] = df_logins.login_time.dt.year
df_logins['month'] = df_logins.login_time.dt.month
df_logins['day'] = df_logins.login_time.dt.day
df_logins['dayOfWeek'] = df_logins.login_time.dt.dayofweek
df_logins['time'] = df_logins.login_time.dt.time
df_logins = df_logins.set_index(df_logins['login_time'])

# create 15 minute intervals
df_logins['agg_15_mins_from'] = df_logins['login_time'].dt.floor('15min')
df_logins['agg_15_mins_to'] =  df_logins['agg_15_mins_from'] + pd.to_timedelta('15min')
df_logins['from'] = df_logins['agg_15_mins_from'].dt.time
df_logins['to'] = df_logins['agg_15_mins_to'].dt.time
df_logins['from'] = df_logins['from'].astype('str')
df_logins['to'] = df_logins['to'].astype('str')
df_logins['interval'] = df_logins['from'] + '-' + df_logins['to']
```

```python
df_logins.head()
```

```python
# Create final analytical dataset for login data
cols = ['year', 'month', 'day', 'dayOfWeek', 'interval']

logins = df_logins[cols]
logins.head()
```

```python
interval_group = logins.groupby(['interval']).dayOfWeek.agg('count').rename('count')
df_interval = pd.DataFrame(interval_group, columns=['count'])
df_interval
```

```python
plt.figure(figsize=(30,20))
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.ylabel('login count')
plt.xlabel('15 minute interval period')


plt.plot(df_interval)
```

We can see that there is a login peak there is a peak between midnight and 1:30am. Likely due to 'Night Owl' activity such as entertainment and nightlife events. There is a trough between 6:30am and 8:30am as these are times when people are usually asleep or commuting to work. I am assuming there aren't a lot of people who would use Ultimate to get to work.

As the day goes on logins increase to another peak at noon. This could be due to people trying to get to lunch appointments although there may be another factor I'm missing here.

As the day continues, activity beings to drop to another trough at 5:15pm which coincides with afternoon rush hour traffic. From here we see logins increase and peaks at 10:45pm which is related to dinner and the start of nightlife activity. 

This graph doesn't segment by day of the week. We should look at this for each day of the week to see if the activity changes based on the day of the week.

```python
df_dayOfWeek = logins.groupby(['dayOfWeek', 'interval']).agg('count')
```

```python
df_dayOfWeek = df_dayOfWeek.drop(['year', 'month'], axis=1)
df_dayOfWeek = df_dayOfWeek.rename(columns={'day': 'count'}).reset_index()
```

```python
df_dayOfWeek.drop(['index'], axis=1)
```

```python
df_pivot = df_dayOfWeek.pivot(index = 'interval',columns = 'dayOfWeek', values = 'count')
df_pivot
```

```python
plt.figure(figsize=(30,20))
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize=15)

plt.plot(df_pivot, label=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.ylabel('login count')
plt.xlabel('15 minute interval')
plt.grid()
plt.legend(loc='best')
```

After grouping counts by day of the week we can see that the counts are slightly higher on friday and alot higher on saturday and sundays. On Saturday and Sunday we see a lot more logins between 1:15am and 5am. 

In terms of data quality, it would be beneficial to have the location of the login as well as the timestamp in that locations timezone. This information can help us determine demand across all regions in one standard timezone to estimate resource usage and also allow us to compare ride demand across equivalent times during the day across all locations.

# Part 2 ‚Äê Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two¬≠way toll, between the two cities causes driver partners to tend to be exclusive to each city. The Ultimate managers of city operations for the two cities have proposed an experiment to encourage driver partners to be available in both cities, by reimbursing all toll costs.
1. What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?
2. Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:
a. how you will implement the experiment
b. what statistical test(s) you will conduct to verify the significance of the
observation
c. how you would interpret the results and provide recommendations to the city
operations team along with any caveats.


## Asnwer

1. We define the sample of driver partners in the experiment as all driver partners that service both Gotham and Metropolis. The metric we would use to measure success for this experiment would be the average number of times a driver crosses the toll booth. I would choose this metric because if a driver would like to service both Cities, they need to cross the toll. If they are not willing to service both cities the drivers would have a low count of the number of times they cross the toll. It doesn't mean they will not, a fare might be worth crossing if the distance travelled is far, so this is why we would like to count the number of times they cross.


2. The experiment would involve splitting partner drivers into two groups, a control group who does not receive the offer and one who receives the ofer to have toll fees reimbursed (Test group). The experiment would be run on weekends since there is similar activity across both cities. We do not want to do the experiment on weekdays since drivers will naturally focus on one city depending on the time of day due to the differences in circadian rhythms.

    Once we run the experiment we will have the number of times a driver crosses the toll on a given day. With this data we can compute the average number of toll crosses by a driver for the control group and the test group. We want to test whether the means of the two groups are equal to zero or if the test group mean is greater than the control group.
    
    To verify the significance of the result that we get, we need to run a permutation test where we randomize the drivers into the test group and control group and then calculate the difference of means, we call this a bootstrap replicate. We can do this a lot of times (ie. 10000) to calculate a lot of replicates to create a distribution of results for the difference of means. We can then compare the actual difference of means we got from the control and test group against this distribution to see what % of bootstrap replicates are equal or more extreme than the difference of means we obtained from the experiment. We conduct a one sided hyptohesis test to see if the % of bootstrap replicates taht are equal or more extreme is 5% or smaller. If it is, then we reject the null hypothesis that the difference of means is 0 and conclude that the offer to reimburse toll costs successfully encourages drivers to service both cities.
    
    We can conduct this experiement across several weekend days depending on the budget we have for this experiemnt. The more days we conduct the experiment the more data we have and the more robust our conclusions. However this will potentially result in more toll fee reimbursements. I would suggest to offer this across 2 full weekends.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
```

```python
df_scatter.columns
```

```python
df.info()
```

```python
# convert retention column to int 

y = df['retention'].astype(int)
```

```python
# Convert boolean columns to int

df['ultimate_black_user'] = df['ultimate_black_user'].astype(int)
```

```python
# get dummy variables for avg_rating_of_driver_bin

driver_rating_dummy = pd.get_dummies(df['avg_rating_of_driver_bin'])
```

```python
driver_rating_dummy
```

```python
# create feature dataframe X

drop_cols = del_cols = ['city', 'signup_date', 'last_trip_date', 'phone', 'retention', 'avg_rating_by_driver', 
                        'avg_rating_of_driver', 'avg_rating_of_driver_bin']

X = df.drop(drop_cols, axis=1)
X.head()
```

```python
# add dummy variables to X dataframe

X = pd.concat([X, driver_rating_dummy], axis=1)
```

```python
X
```

From a modeling perspective we will use a logistic regression to start as this model allows for better interpretability in terms of the impact of the features on the prediction. We can obtain better inference with this model relative to a black box model.

```python
# create training set and test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# instantiate logistic regression model

lr = LogisticRegression(solver='liblinear')
```

```python
# fit the clasifier to the training data

lr.fit(X_train, y_train)

# predict on test set

y_pred = lr.predict(X_test)
```

```python
# Compute and print the confusion matrix and classification report

print(confusion_matrix(y_test, y_pred, labels=[0,1]))
```

```python
print(classification_report(y_test, y_pred))
```
